```python
#librerias nesesarias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Datos 1

```python
#cargando los datos
data = pd.read_excel("datos_1.xlsx")
```

### Parte 1, fig 3 y fig 7

Parte 1 del experimento, medicion de la componente z para distintas distancias
entre la bobinas, ver figura 7 de la guia


```python
#distancia R
Z_R = data.iloc[6:47,1].astype(float).to_numpy()
Bz_R = data.iloc[6:47,2].astype(float).to_numpy()
#distancia R/2
Z_Rmed=data.iloc[6:47,4].astype(float).to_numpy()
Bz_Rmed = data.iloc[6:47,5].to_numpy()
#distancia 2R
Z_2R =data.iloc[6:47,7].astype(float).to_numpy()
Bz_2R = data.iloc[6:47,8].to_numpy()
```

```python
plt.plot(Z_R,Bz_R,'-o',label='Distancia R')
plt.plot(Z_Rmed,Bz_Rmed,'-o',label='Distancia R/2')
plt.plot(Z_2R,Bz_2R,'-o',label='Distancia 2R')
plt.xlabel('z')
plt.ylabel('B_z')
plt.grid(linewidth=1)
plt.legend()
#plt.figure(figsize=(1,30))
plt.show()
```

### Parte 2, fig 4 y fig 8

Parte 2 del experimento, medicion de la componente z a una distancia R entre las bobinas, 
pero para distintas distancias radiales r, ver figura 8 de la guia

```python
Z = data.iloc[53:73,1].astype(float).to_numpy()

r_0 = data.iloc[53:73,2].astype(float).to_numpy()   #r=0mm
r_10=data.iloc[53:73,3].astype(float).to_numpy()    #r=10mm
r_14 = data.iloc[53:73,4].to_numpy()                #r=14mm
r_16 =data.iloc[53:73,5].astype(float).to_numpy()   #r=16mm
```

```python
plt.plot(Z,r_0,'-o',label='r=0mm')
plt.plot(Z,r_10,'-o',label='r=10mm')
plt.plot(Z,r_14,'-o',label='r=14mm')
plt.plot(Z,r_16,'-o',label='r=16mm')
plt.grid()
plt.xlabel('zmm')
plt.ylabel('B_z')
plt.legend()
```

### Parte 3, fig 5 y fig 9

Parte 3 del experimento, medicion de la componente r a una distancia R entre las bobinas, 
pero para distintas distancias radiales r, ver figura 9 de la guia.

```python
Z = data.iloc[79:97,1].astype(float).to_numpy()

Br_r_0 = data.iloc[79:97,2].astype(float).to_numpy()   #r=0
Br_r_10=data.iloc[79:97,3].astype(float).to_numpy()    #r=10
Br_r_14 = data.iloc[79:97,4].to_numpy()                #r=14
Br_r_16 =data.iloc[79:97,5].astype(float).to_numpy()   #r=16
```

```python
plt.plot(Z,Br_r_0,'-o',label='r=0mm')
plt.plot(Z,Br_r_10,'-o',label='r=10mm')
plt.plot(Z,Br_r_14,'-o',label='r=14mm')
plt.plot(Z,Br_r_16,'-o',label='r=16mm')
plt.grid()
plt.legend()
plt.xlabel('zmm')
plt.ylabel('B_r')
plt.show()
```

### Parte 4, fig 10

Parte 4 del experimento, medicion de la componente r a una distancia R entre las bobinas, 
pero para distintas distancias radiales r en Z=0, ver figura 10 de la guia.

```python


Z_0 = data.iloc[102:122,1].astype(float).to_numpy()

Br_1 = data.iloc[102:122,2].astype(float).to_numpy()
Br_2=data.iloc[102:122,3].astype(float).to_numpy()
```

```python
plt.plot(Z_0,Br_1,'-o',label='Bobina 1')
plt.plot(Z_0,Br_2,'-o',label='Bobina 2')
plt.grid()
plt.legend()
plt.xlabel('r')
plt.ylabel('B_r')
```
