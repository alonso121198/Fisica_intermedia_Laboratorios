### Parte 1, fig 3 y fig 7

Parte 1 del experimento, medicion de la componente z para distintas distancias
entre la bobinas, ver figura 7 de la guia


```python
#distancia R
Z_R = data.iloc[6:47,1].astype(float).to_numpy()
Bz_R = data.iloc[6:47,2].astype(float).to_numpy()
#distancia R/2
Z_Rmed=data.iloc[6:47,4].astype(float).to_numpy()
Bz_Rmed = data.iloc[6:47,5].to_numpy()
#distancia 2R
Z_2R =data.iloc[6:47,7].astype(float).to_numpy()
Bz_2R = data.iloc[6:47,8].to_numpy()
```

```python
import lmfit # este sirve para no linear "Non-Linear Least-Squares Minimization and Curve-Fitting"
```

# Datos 1 Experimento 1

```python
#cargando los datos
data = pd.read_excel("Datos_EXP_2_A.xlsx")
# display(data)

# eliminamos las filas y columnas que no queremos
data = data.drop([0,2],axis=0) # elimina las 4 primeras filas
data = data.drop(data.columns[[0,3]],axis=1) # elimina las 4 primeras columnas



# Lo que hacemos ahora es resetear indices para no tener problemas luego en la elaboracion
data1 = data.reset_index(drop=True)
data1
```

```python
# El for va por todas las listas y en paralelo va de "A" hacia "L" indicado en la tabla de datos
linear = lmfit.models.LinearModel()


plt.plot(data1.iloc[1:,0],data1.iloc[1:,1],"*",label="a")
plt.xlabel('Intensidad(A)',fontsize=20)
plt.ylabel('Campo Magnetico(mT)',fontsize=20)
plt.grid(True)
plt.show()

print("#"*100)
modelado1 = linear.fit(list(data1.iloc[1:,1]), x=list(data1.iloc[1:,0]) ) 
fig, gs = modelado1.plot()
fig.axes[1].grid()
fig.axes[1].set_xlabel('Intesidad(A)',fontsize=20)
fig.axes[1].set_ylabel('Campo Magnetico (mT)',fontsize=20)
fig.axes[0].set_title("data",fontsize=20)
plt.show()
display(modelado1)

print("#"*100)
angulos = np.tan(data1.iloc[1:10,3].apply(pd.to_numeric,errors='coerce').to_numpy()*(2*np.pi/360) )
plt.plot(angulos,data1.iloc[1:10,2],"*",label="a")
plt.xlabel(r'$Tan(\alpha)$',fontsize=20)
plt.ylabel(r'$I_{H}$ ',fontsize=20)
plt.grid(True)
plt.show()



print("#"*100)
angulos = np.tan(data1.iloc[1:10,3].apply(pd.to_numeric,errors='coerce').to_numpy()*(2*np.pi/360) )
modelado2 = linear.fit(list(data1.iloc[1:10,2]), x=angulos ) 
fig, gs = modelado2.plot()
fig.axes[1].grid()
fig.axes[1].set_xlabel(r'$Tan(\alpha)$',fontsize=20)
fig.axes[1].set_ylabel(r'$I_{H}$ ',fontsize=20)
fig.axes[0].set_title("data",fontsize=20)
plt.show()
display(modelado2)

plt.show()
```

# Luego el campo Magnetico horizontal esta dado por 
$$ B_{E}^{h} = m1*m2 $$ 
Donde m1 y m2 son las pendientes de las dos graficas en la interpolacion respectivamente

```python
m1 = 0.62832840
m2 = 0.04709181
Bh = m1*m2
Bh
```

## Luego el campo magnetico terrestre esta dado por ##
$$ B_{E} = B_{E}^{h}*sec(\theta)  $$ 

```python
teta = (28+32)/2
B = Bh/np.cos(teta*np.pi/180)
B
```

```python
## en microTesla
B*10**3
```
